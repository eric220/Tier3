**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#get col names except first, transpose df
col_name = df.columns[:1]
cols = df[col_name]
df_t = df.T.reset_index()
df_t.columns = df_t.iloc[0]
df_t = df_t.drop(0)
df_t.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#rename to bourogh col
df_t.rename(columns={ df_t.columns[0]: "Bourough" }, inplace = True)
```

```python
#List from wikipedia, change City of Westminster to Westminster, get not bouroughs list
bouroughs = ['Westminster', 'Kensington & Chelsea', 'Hammersmith & Fulham', 'Wandsworth', 'Lambeth',
             'Southwark', 'Tower Hamlets', 'Hackney', 'Islington', 'Camden', 'Brent', 'Ealing', 'Hounslow',
             'Richmond upon Thames', 'Kingston upon Thames', 'Merton', 'Sutton', 'Croydon', 'Bromley', 'Lewisham',
             'Greenwich', 'Bexley', 'Havering', 'Barking & Dagenham', 'Redbridge' ,'Newham', 'Waltham Forest',
             'Haringey', 'Enfield', 'Barnet', 'Harrow', 'Hillingdon']
not_bouroughs = [x for x in df_t.Bourough[:] if x not in bouroughs]
```

```python
#drop not bouroughs form df, assert len df == len bouroughs 
def drop_not_bouroughs(tdf, col, to_drop):
    return df_t[~df_t[col].isin(to_drop)]
df_t = drop_not_bouroughs(df_t, 'Bourough', not_bouroughs)  
assert len(df_t) == len(bouroughs)
```

```python
#set index to bourough and drop 
df_t.set_index(df_t['Bourough'], inplace = True)
df_t.drop('Bourough', inplace = True, axis = 1)
df_t.head(3)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
#plot price increase over time
df = df_t.T
df.plot(legend = False, figsize = (10,7), title = 'Price increase over time')
plt.ylabel('Hundred Thousands')
plt.xlabel('');
```

```python
#plot greatest overall increase
top_num = 5
df_t.sort_values(by = df_t.columns[-1:][0], ascending = False).head(top_num).T.plot(figsize = (10,7),
                                                                                    title = 'Top 5 Bourough Prices')
plt.ylabel('Hundred Thousands')
plt.xlabel('');
```

```python
#get percent change and total change
df_t['pct_chg'] = ((df_t[df_t.columns[-1:]].values-df_t[df_t.columns[:1]].values)/df_t[df_t.columns[:1]].values)*100
df_t['total_chg'] = df_t[df_t.columns[-2:-1]].values - df_t[df_t.columns[:1]].values
```

```python
#plot total change
df_t['total_chg'].plot(kind = 'bar', color = 'lightblue', figsize = (10,7), title = 'Total Price Change');
```

```python
#plot percent change
df_t['pct_chg'].plot(kind = 'bar', color = 'lightblue', figsize = (10,7), title = 'Percent Change');
```

```python
#sort by starting price descending
df_t.sort_values(by = df_t.columns[:1][0], ascending = False)[['total_chg', 'pct_chg']].head()
```

```python
#sort by percent change descending
df_t.sort_values(by = 'pct_chg', ascending = False)[['total_chg', 'pct_chg']].head()
```

```python
sorted_df = df_t.sort_values(by = df_t.columns[:1][0])
sorted_df.head()
```

```python
#plot starting price by percent change
sorted_df = df_t.sort_values(by = df_t.columns[:1][0])
x = sorted_df.index
y = np.array(sorted_df[sorted_df.columns[:1]]).reshape(-1)
s = np.array((sorted_df['pct_chg']/np.array(np.max(sorted_df['pct_chg'])))*1000)
plt.figure(figsize=(10,7))
ax = sn.scatterplot(x=x, y=y, s = s, hue = np.array((sorted_df['pct_chg'])), legend = True, palette= sn.color_palette('magma_r', 8, as_cmap=True))
plt.title('Starting Price and Percent Gain')
plt.xticks(rotation=85)
plt.show()
```
